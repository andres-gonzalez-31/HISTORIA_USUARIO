# Extracción de Datos desde PostgreSQL y Exportación a CSV

En este notebook se realiza la conexión a la base de datos PostgreSQL, se extraen las tablas **ventas**, **clientes** y **productos**, y se exportan sus datos a archivos CSV como respaldo.


Importacion de librerias

```python
import pandas as pd
from sqlalchemy import create_engine
```

## Configuración de la conexión a PostgreSQL

Se establece la conexión utilizando SQLAlchemy.  
Actualizar las credenciales según la configuración local.


```python
usuario = "postgres"
contrasena = "Andy310823*"   # Cambiar según corresponda
host = "localhost"
puerto = "5432"
base_datos = "ventas_UH_RIWI"

engine = create_engine(f"postgresql://{usuario}:{contrasena}@{host}:{puerto}/{base_datos}")
engine.connect()
```

## Extracción de tablas

Se consultan y cargan en DataFrames las tablas:
- ventas  
- clientes  
- productos


```python
df_ventas = pd.read_sql("SELECT * FROM fact_ventas", engine)
df_clientes = pd.read_sql("SELECT * FROM dim_tipos_clientes", engine)
df_productos = pd.read_sql("SELECT * FROM dim_productos", engine)

df_ventas.head(), df_clientes.head(), df_productos.head()
```

## Exportación de respaldos a CSV

Se generan archivos CSV con el contenido actual de cada tabla.


```python
df_ventas.to_csv("respaldo_ventas.csv", index=False)
df_clientes.to_csv("respaldo_clientes.csv", index=False)
df_productos.to_csv("respaldo_productos.csv", index=False)

print("Respaldo generado correctamente.")
```
